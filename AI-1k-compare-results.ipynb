```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def render_mpl_table(data, row_labels, max_indices, col_width=3.0, row_height=0.625, font_size=14,
                     header_color='#40466e', row_colors=['#f1f1f2', 'w'], edge_color='w',
                     bbox=[0, 0, 1, 1], header_columns=0,
                     ax=None, **kwargs):
    if ax is None:
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height])
        fig, ax = plt.subplots(figsize=size)
        ax.axis('off')

    mpl_table = ax.table(cellText=data.values, bbox=bbox, colLabels=data.columns, rowLabels=row_labels, **kwargs)

    mpl_table.auto_set_font_size(False)
    mpl_table.set_fontsize(font_size)

    for k, cell in mpl_table._cells.items():
        cell.set_edgecolor(edge_color)
        if k[0] == 0 or k[1] < header_columns:
            cell.set_text_props(weight='bold', color='w')
            cell.set_facecolor(header_color)
        else:
            cell.set_facecolor(row_colors[k[0] % len(row_colors)])
            if k in max_indices:
                cell.set_text_props(weight='bold')

    return 

def calculate_average_metrics(experiment_names, metric_names):
    results = []

    for experiment_name in experiment_names:
        experiment_values = []
        for metric_name in metric_names:
            metric_value = load_split_best_metric_from_experiment(experiment_name, metric_name)
            mean_value = np.mean(metric_value)
            std_value = np.std(metric_value)
            experiment_values.append((mean_value, std_value))
        results.append(experiment_values)

    return results

def create_results_table(experiment_names, metric_names, output_image_path):
    results = calculate_average_metrics(experiment_names, metric_names)

# plt.plot(['CLIP-IQA'], hps_values_4, marker='o', linestyle='-', markersize=8, color='red')

plt.xlabel('Similarity Metrics')
    table_data = []
    max_indices = []

    for i, experiment_name in enumerate(experiment_names):
        row_data = []
        for j, metric_name in enumerate(metric_names):
            mean_value, std_value = results[i][j]
            row_data.append(f"{mean_value:.4f} ({std_value:.4f})")
        table_data.append(row_data)

        if metric_name == "mse":
            best_index = min(range(len(row_data)), key=row_data.__getitem__)
        else:
            best_index = max(range(len(row_data)), key=row_data.__getitem__)
        max_indices.append((i+1, best_index))

    col_labels = [f"{metric_name.capitalize()} Mean (Std)" for metric_name in metric_names]
    row_labels = experiment_names

    ax = render_mpl_table(pd.DataFrame(table_data, columns=col_labels, index=row_labels), row_labels=row_labels, 
                          max_indices=max_indices, header_columns=0, col_width=4.0)
    plt.savefig(output_image_path, bbox_inches='tight')
    plt.show()
    plt.close()


    


output_image_path = os.path.join(plot_dir, "results_table.png")
create_results_table(experiment_names, metric_names, output_image_path)
```
